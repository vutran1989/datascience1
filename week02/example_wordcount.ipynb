### Open and read the csv file

```python
file = open("kieu.csv")
```

```python
s = file.read()
```

### Clean the data and split the string into multiple lines

```python
import re
kieu_list = re.sub("[0-9.,:?!'\";“\\-”]", "", s).split("\n")
```

```python
kieu = [line.strip() for line in kieu_list]
kieu = list(filter(lambda x: len(x) > 0, kieu))
```

### Word count using dictionary

```python
kieu_all_words = " ".join(kieu).lower().split()
```

#### 1-gram

```python
count = {}
for word in kieu_all_words:
    count[word] = count.get(word, 0) + 1
```

```python
sorted(count.items(), key=lambda x: x[1], reverse=True)
```

#### 2-gram

```python
kieu_2words = [kieu_all_words[i] + " " + kieu_all_words[i + 1] for i, _ in enumerate(kieu_all_words) if i < len(kieu_all_words) - 1]
```

```python
count_2 = {}
for word in kieu_2words:
    count_2[word] = count_2.get(word, 0) + 1
```

```python
count_2_sorted = sorted(count_2.items(), key=lambda x: x[1], reverse=True)
```

```python
count_2_sorted
```

### Word count using itertools groupby

```python
from itertools import groupby
```

#### 1-gram

```python
words_map = sorted([(word, 1) for word in kieu_all_words])
```

```python
sorted([(key, len(list(value))) for key, value in groupby(words_map, lambda x: x[0])], key=lambda x: x[1], reverse=True)
```

#### 2-gram

```python
words_map = sorted([(word, 1) for word in kieu_2words])
```

```python
sorted([(key, len(list(value))) for key, value in groupby(words_map, lambda x: x[0])], key=lambda x: x[1], reverse=True)
```
